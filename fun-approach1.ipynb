```python
raw_data=pd.read_csv('../input/function/fun.csv')
```

```python
raw_data.head()
```

```python
X=raw_data.drop(['x','f','rand','f_r'], axis=1)
X=X.drop([0])
y=raw_data['f_r']
y=y.drop([0])
```

```python
X=X.to_numpy()
y=y.to_numpy()
```

```python
type(y)
```

```python
from sklearn import linear_model
regr=linear_model.Lasso(fit_intercept= False)
regr.get_params()
```

```python
regr.fit(X,y)
```

```python
print (regr.coef_, regr.intercept_)
```

```python
alphas = np.arange(1, 100, 5)
lassoCV_regressor = linear_model.LassoCV(alphas=alphas, cv=3)
lassoCV_regressor.fit(X, y)
```

```python
%pylab inline
plot(lassoCV_regressor.alphas_, lassoCV_regressor.mse_path_.mean(1)); 
#dl'a kazhdoj stroki po stolbtsam
xlabel('alpha');
ylabel('усредненный_ по строкам MSE');
```
