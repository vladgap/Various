### Transform list to set_weights

```python
def arrange_weight(layers, w_list):
    w=[]
    for i in [0,1]:
        layer=[]
        for j in range(layers[i+1]):
            neuron=w_list[(layers[i]+1)*j+i*(layers[0]+1)*layers[1]:(layers[i]+1)*(j+1)+i*(layers[0]+1)*layers[1]]
            layer.append(neuron)
        w.append(layer)
    return w    
```

### Creating list of lists of MSEs

get2_MSE

```python
# use MultiLayerNeuron instead of Regres for saving time
layers1=[2,2,1]
mln1=MultiLayerNeuron(layers=layers1)
mln1.set_activ('mrelu', param=0.01)
mln1.set_output_activ('linear')
```

```python
def get2_MSE1(w):
    w_arr=arrange_weight(layers1, w)
    mln1.set_weights(w_arr)
    pred=[]
    for i in range(len(X)): # rows in X
        y=[]
        for j in range(layers1[1]): # num. of neurons in hidden layer
            a=np.dot(np.append(X[i],1),mln1.network[1][j].weights)
            if a>=0:
                y.append(a)
            else:
                y.append(mln1.param*a)
        y.append(1) # the bias
        b=np.dot(y,mln1.network[2][0].weights)
        pred.append(b)
#     print (pred)    
    MSE=((2*(pred-y1))**2).sum()/len(y1)
    return MSE
```

```python
import sys, os

# Disable
def blockPrint():
    sys.__stdout__ = sys.stdout
    sys.stdout = open(os.devnull, 'w')

# Restore
def enablePrint():
    sys.stdout = sys.__stdout__

print ('This will print')

blockPrint()
print ("This won't")

enablePrint()
print ("This will too")
```

```python
def get_final_MSE(w):
    # w=a.iloc[13766]
    arr_w=arrange_weight(layers1, w)
    reg_test=Regres(layers=layers1)
    reg_test.set_hidden_activ('mrelu', param=0.01)
    reg_test.set_weights(arr_w)
    reg_test.fit(np.array(X), y1, epochs=2000, eta=0.01)
    MSE_test=[get2_MSE1(w), reg_test.MSE_history[0], reg_test.MSE_history[-1]]
    return MSE_test
```
