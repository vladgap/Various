```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python
data=np.array([[90,0,225,31.1],
[117,0,208,31.1],
[140,0.7,211,31.1],
[120,0,234,22.12],
[120,3.8,179,22.12],
[120,3.6,157,22.12],
[120,0,198,22.12],
[140,0.7,134,22.12],
[140,5.2,148,22.12],
[90,0,179,22.12],
[140,3,196,22.12],
[120,0,234,24.11],
[120,3.5,179,24.11],
[120,3.5,157,24.11],
[120,0,198,24.11],
[140,0,134,24.11],
[140,5,148,24.11]])
```

```python
data_4D=pd.DataFrame(data, columns=('Temp','Pres','TOC','day'))
data_4D
```

```python
fig = go.Figure()

x,y,z=data_4D.Temp,data_4D.Pres,data_4D.TOC
fig.add_trace(go.Scatter3d(x=x, y=y, z=z, mode='markers',marker = dict(color=data_4D.day, colorscale=['green','yellow','blue']) ,text=data_4D.day, name='4D'))

y_proj=0
for i in range(len(x)):
    x_line=[x[i],x[i]]
    y_line=[y_proj,y[i]]
    z_line=[z[i],z[i]]
    fig.add_trace(
        go.Scatter3d(x=x_line, y=y_line, z=z_line, mode='lines',
                    line = dict(color='black', width=2, dash='dash'),
                    showlegend=False)
    )
z_proj=100
for i in range(len(x)):
    x_line=[x[i],x[i]]
    y_line=[y[i],y[i]]
    z_line=[z_proj,z[i]]
    fig.add_trace(
        go.Scatter3d(x=x_line, y=y_line, z=z_line, mode='lines',
                    line = dict(color='black', width=2, dash='dash'),
                    showlegend=False)
    )

fig.update_layout(
    margin=dict(l=0, r=0, b=0, t=0),
    scene = dict(
        xaxis_title='Temp',
        yaxis_title='Pres',
        zaxis_title='TOC',
#         xaxis = dict(nticks=5, range=[15,7]),
#         yaxis = dict(nticks=10, range=[13,0]),
        zaxis = dict(nticks=10, range=[100,240]),
    ),
    width=650,height=500,
#     scene_camera=dict(eye=dict(x=-.5, y=3, z=0.)),
)    
fig.show()
```
